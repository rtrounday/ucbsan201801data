```python
# Load Dependencies
from bs4 import BeautifulSoup as bs
from splinter import Browser
import os
import requests
import pandas as pd
```

```python
#Confirm working directory
cwd = os.getcwd()
print (cwd)
```

```python
# URL of page to be scraped for Mar News
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
response = requests.get(url)
soup = bs(response.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
news_results = soup.find_all('div', class_="slide")
```

```python
print (news_results)
```

```python
for result in news_results:
    # Error handling
    try:
        # Identify and return title of listing
        news_title = result.find('div', class_="content_title").text
        # Identify the text associated with title
        news_teaser = result.find('div', class_="rollover_description_inner").text
        # Identify and return link of listing
        link = result.a["href"]

        # Print results only if title, news and link is available
        if (news_title and news_teaser and link):
            print('-------------')
            print(news_title)
            print(news_teaser)
            print(link)
    except Exception as e:
        print(e)
```
